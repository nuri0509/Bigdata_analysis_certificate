```python
#1번. 주어진 데이터를 이용하여 이동시간(출동시간과 도착시간의 차이)이 평균적으로 가장 오래걸린 구급대의 이동시간을 구하시오
#     (분 단위로 환산 및 반올림하여 정수로 출력) - 데이터 : data_6_1_1.csv
```

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_1.csv')
df1.head(10)
```

```python
df1.info()
```

```python
df1['출동시간'] = pd.to_datetime(df1['출동시간'])
df1['도착시간'] = pd.to_datetime(df1['도착시간'])
```

```python
df1['이동시간'] = df1['도착시간'] - df1['출동시간']
df1['이동시간'] = df1['이동시간'].dt.total_seconds() / 60
print(df1)
```

```python
group_mean = df1.groupby('구급대')['이동시간'].mean()
print(group_mean)
```

```python
result = round(group_mean.max())
print(result)
```

```python
#2번. 주어진 데이터를 이용하여 교수 1인당 맡고있는 학생수가 가장 많은 대학교를 찾고, 그 대학교의 전체 교수의 수를 구하시오
#      (정수로 출력) - 데이터 : data_6_1_2.csv
```

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_2.csv')
print(df)
```

```python
df['학생'] = df.iloc[:, 2:].sum(axis=1)
print(df.head())
```

```python
df['학생/교수'] = df['학생'] / df['교수']
print(df.head())
```

```python
max_id = df['학생/교수'].idxmax()
result = int(df.loc[max_id, '교수'])
print(result)
```

```python
#3번  주어진 데이터를 이용하여 연도별 총 범죄 건수의 월 평균값을 구하고, 그 값이 가장 큰 연도의 총 범죄건수의 월평균을 집계하시오
#     반올림하여 정수로 출력 - 데이터 : data_6_1_3.csv
```

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_1_3.csv')
print(df)
```

```python
df['총범죄건수'] = df.iloc[:, 1:-1].sum(axis=1)
print(df.head())
```

```python
df['연도'] = df['날짜'].str[:4]
print(df.head())
```

```python
group_mean = df.groupby("연도")['총범죄건수'].mean()
print(group_mean)
```

```python
result = round(group_mean.max())
print(result)
```
