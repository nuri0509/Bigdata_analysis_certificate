```python
# 문제 32
# 근속연수 변수의 결측치를 각 부서별 같은 평가등급의 평균 근속연수로 채우고, 근속연수의 평균을 구하시오.
# (단, 근속연수의 결측치, 평균값은 소수점을 버리고 정수로 채우시오.)

### 복사영역 ### 
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "부서": ["marketing"] * 9 + ["it"] * 9,
    "평가등급": ["A", "A", "A", "B", "B", "B", "C", "C", "C", 
             "A", "A", "A", "B", "B", "B", "C", "C", "C"],
    "근속연수": [9, None, 12, 14, None, 17, 19, None, 22, 
             2, None, 5, 5, None, 8, 8, None, 10]
})
### 복사영역 ### 
df
```

```python
# 문제 37
# mpg값이 25이상이면 high, 15이상 25 미만이면 medium, 15미만이면 low로 범주화하고
# mpg값이 high인 그룹과 low인 그룹의 disp 표준편차의 차이를 절대값으로 구하시오.

df = pd.read_csv('./data/mtcars.csv')
df.head()
```

```python
df['mpg_group'] = df['mpg'].apply(lambda x : 'high' if x>=25  else 'medium' if x>=15 else 'low')
df.head()
```

```python
# 문제 40
# 다음은 학생별(학생ID) 선택과목이 점수, 과목별 강의실 정보에 대한 데이터이다.
# 학점당 평균 점수가 가장 높은 학생의 학생 ID를 구하시오.
# * 학점당 평균 점수는 가중점수의 총합을 전체 학점수로 나눈 값이다.
# * 가중 점수는 과목별 점수x과목별 학점수로 계산한다.

## 복사 영역 ##
import pandas as pd
df = pd.DataFrame({
    '학생ID': [201, 202, 203, 204, 205],
   # '선택과목': ['수학', '영어', '과학', '수학', '영어'],
    '수학_점수': [88, 92, 85, 91, 87],
    '영어_점수': [78, 81, 79, 75, 83],
    '과학_점수': [89, 85, 90, 88, 84],
   # '수학_강의실': ['101호']*5,
   # '영어_강의실': ['102호']*5,
   # '과학_강의실': ['103호']*5
})
# 학점 정보 추가
df['수학_학점수'] = 3
df['영어_학점수'] = 2
df['과학_학점수'] = 1

## 복사 영역 ##
df
```

```python
df['전체학점수'] = df['수학_학점수'] + df['영어_학점수'] + df['과학_학점수']
df['학점당평균'] = df['total'] / df['전체학점수']
df
```

```python
# 문제 41
# 다음은 어느 회사의 지역별(region) 성별에 따른 제품별 판매량 데이터이다.
# 지역별 성별에 따른 총 판매량을 구하고, 성별간 총 판매수 차이가 큰 지역의 지역명을 구하시오.
# (단, 총 판매수는 product1과 product2를 더한 값이다.)

## 복사 영역 ##
import pandas as pd
data = {
    'region': ['a']*4 + ['b']*4 + ['c']*4 + ['d']*4 + ['e']*4,
    'gender':  [1, 2, 1, 2,   2, 1, 2, 1,   1, 1, 2, 2,   2, 1, 1, 2,   1, 2, 2, 1],
    'product1': [
        1000, 2000, 2500, 3000,   # a
        4000, 120, 250, 350,      # b
        5000, 6000, 5500, 5800,   # c
        7000, 7200, 7100, 7300,   # d
        8000, 8200, 8100, 8300    # e
    ],
    'product2': [
        3000, 1500, 1500, 2000,   # a
        3500, 350, 290, 1000,     # b
        4000, 4500, 4200, 4300,   # c
        6500, 6700, 6600, 6800,   # d
        7500, 7700, 7600, 7800    # e
    ]
}
df = pd.DataFrame(data)
## 복사 영역 ##
df.head(10)
```

```python
# 문제 42
# 다음은 어느 회사의 지역별(region) 성별에 따른 제품별 판매량 데이터이다.
# product1 변수의 결측치를 각 region별 같은 성별의 평균 값으로 채우고,
# 지역별 product1과 product2 판매수의 차이가 가장 큰 지역을 찾으시오.

## 복사 영역 ##
import pandas as pd
data = {
    'region': ['a']*4 + ['b']*4 + ['c']*4 + ['d']*4 + ['e']*4,
    'gender':  [1, 2, 1, 2,   2, 1, 2, 1,   1, 1, 2, 2,   2, 1, 1, 2,   1, 2, 2, 1],
    'product1': [
        1000, 2000, 2500, 3000,   # a
        4000, 120, 250, 350,      # b
        5000, 6000, 5500, 5800,   # c
        7000, 7200, 7100, 7300,   # d
        8000, 8200, 8100, 8300    # e
    ],
    'product2': [
        3000, 1500, 1500, 2000,   # a
        3500, 350, 290, 1000,     # b
        4000, 4500, 4200, 4300,   # c
        6500, 6700, 6600, 6800,   # d
        7500, 7700, 7600, 7800    # e
    ]
}
df2 = pd.DataFrame(data)
df = df2.copy()
# 각 지역에서 첫 번째 등장하는 인덱스를 기준으로 결측치 처리
for region in ['a', 'b', 'c']:
    idx = df[df['region'] == region].index[0]
    df.loc[idx, 'product1'] = np.nan

## 복사 영역 ##
df.head(10)
```

```python
df['product1'] = df['product1'].fillna(df.groupby(['region','gender'])['product1'].transform('mean'))
df.head()
```

```python
pivot1 = pd.pivot_table(data=df, 
                        index='region',
                        values=['product1','product2'],
                        aggfunc='sum')
pivot1
```

```python
pivot1['result'] = abs(pivot1['product1'] - pivot1['product2'])
pivot1
```

```python
pivot1['result'].idxmax()
```
