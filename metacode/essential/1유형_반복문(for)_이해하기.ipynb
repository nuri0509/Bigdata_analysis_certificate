```python
# list comprehension(리스트 컴프리헨션)
# 문법별 차이 확인해보기

# (1) for문으로 print 반복
for i in range(3):
    print(i)

# (2) for + append 활용
num = [] # 빈 리스트를 먼저 생성
for i in range(3):
    num.append(i)
print(num)   # 반복문을 통해 빈 리스트에 데이터를 추가하기

# (3) 리스트 컴프리헨션 사용
result = [i for i in range(3)]
print(result)
```

```python
# 실전 예제 풀이

## 데이터 생성
df1 = pd.DataFrame({'연도':[2022, 2022, 2023, 2023, 2024, 2024],
                    '구분':['판매수', '등록수', '판매수', '등록수', '판매수', '등록수'],
                    '항목1':[95,97,48,43,18,35],
                    '항목2':[12, 11, 20, 87, 4, 82],
                    '항목3':[55, 64, 17, 83, 19, 42]})

df2 = df1[df1['구분'] == '판매수']
df3 = df1[df1['구분'] == '등록수']
df = pd.merge(df2, df3, on='연도', suffixes=('_판매수', '_등록수'))
df

## 데이터 생성
df.head()
```

```python
# 예제 1 : for 문을 사용해서 항목별 판매비율 열을 추가해보자
# 판매비율은 판매수를 등록수로 나눈값이다.(판매비율 = 판매수/등록수)

# 만약 각 항목별로 판매비율에 대한 파생변수를 추가하려면, 시간이 너무 오래걸림.
# df['항목1_판매비율'] = df['항목1_판매수'] / df['항목1_등록수'] 
# df['항목2_판매비율'] = df['항목2_판매수'] / df['항목2_등록수'] 
# df['항목3_판매비율'] = df['항목3_판매수'] / df['항목3_등록수'] ...
```

```python
# for 문을 활용해서 반복 작업을 간단히 처리하기
for i in range(1,4):
    df[f'항목{i}_판매비율'] = df[f'항목{i}_판매수'] / df[f'항목{i}_등록수']

df
```

```python
## 생각해보기 : 위와 같은 작업은 왜 리스트 컴프리헨션은 안될까?
# --> 리스트 컴프리헨션은 리스트를 생성할 때 사용!
# --> 위작업은 다수의 파생변수를 생성하는 작업이기 때문에 적합하지 않음. 
```

```python
# 예제 2 : 판매수, 등록수, 판매비율이 가장 큰 항목을 찾아보기
# 위 데이터셋에서 판매수가 가장 큰 항목이 무엇인지 찾고 싶다면?
result1 = df[['항목1_판매수', '항목2_판매수','항목3_판매수']].idxmax(axis=1)
result2 = df[[f'항목{i}_판매수' for i in range(1,4)]].idxmax(axis=1)

sample = [f'항목{i}_판매수' for i in range (1,4)]

print(result1)
print(result2)
print(sample)
```

```python
# 만약 파생변수로 추가하고 싶다면?
df['최대판매항목'] = result1
df
```

```python
# 위 데이터셋에서 등록수가 가장 큰 항목이 무엇인지 찾고 싶다면?
df[[f'항목{i}_등록수' for i in range(1,4)]].idxmax(axis=1) # 열 이름 찾기
# df[['항목1_등록수','항목2_등록수','항목3_등록수]].idxmax(axis=1)
```

```python
# 위 데이터셋에서 판매비율이 가장 큰 항목이 무엇인지 찾고 싶다면?
df[[f'항목{i}_판매비율' for i in range(1,4)]].idxmax(axis=1)
# df[['항목1_판매비율', '항목2_판매비율', '항목3_판매비율']].idxmax(axis=1)
```

```python
# 예제 3) 풀이 : for + append + 문자열 더하기
## 데이터 생성

df = pd.DataFrame({
'학생ID': [201, 202, 203, 204, 205],
'선택과목': ['수학','영어','과학','수학','영어'],
'수학_점수': [88, 92, 85, 91, 87],
'영어_점수': [78, 81, 79, 75, 83],
'과학_점수': [89, 85, 90, 88, 84],
'수학_강의실': ['101호']*5,
'영어_강의실': ['102호']*5,
'과학_강의실': ['103호']*5
})
df
```

```python
# 미리 최종 결과(output)를 생각해보면 4개의 컬럼으로 구성된다.
# '학생별로 선택한 과목의 점수와 강의실 추출' 
# 학생ID / 선택과목 / 선택과목점수 / 선택과목 강의실 
```

```python
# for문을 활용해서 값을 추출하기

score = []  # 선택과목의 '점수'를 담을 리스트

for i in range(len(df)): 
    row = df.iloc[i]
    과목 = row['선택과목']
    점수 = row[과목 + '_점수']  # 문자열 더하기, f-string : row[f'{과목}_점수']
    score.append(점수)

df['점수'] = score   # 파생변수 추가
df
```

```python
room = []  # 선택과목의 '강의실'을 담을 리스트

for i in range(len(df)):
    row = df.iloc[i]
    과목 = row['선택과목']
    강의실 = row[과목 + '_강의실']
    room.append(강의실)

df['강의실'] = room  # 파생변수 추가
df
```

```python
# 새 컬럼으로 추가
df['선택과목_점수'] = score
df['선택과목_강의실'] = room
df
```

```python
result = df[['학생ID', '선택과목', '선택과목_점수', '선택과목_강의실']]
result
```

```python
# 예제 4) 풀이 2 : for + append + 문자열 더하기
df = pd.DataFrame({
'환자ID': [301, 302, 303, 304, 305],
'복용약물': ['X','Y','Z','X','Y'],
'X_용량': [100, 150, 100, 120, 130],
'Y_용량': [200, 180, 190, 210, 170],
'Z_용량': [300, 310, 290, 305, 295],
'X_복용시간': ['아침']*5,
'Y_복용시간': ['점심']*5,
'Z_복용시간': ['저녁']*5
})
### 데 이 터 생 성 ###
df
```

```python
# for 문을 활용해서 값을 추출하기

# 결과 담을 리스트
dosage_list = []  # 용량만
time_list = []  # 복용시간만

# 1. 용량 리스트 만들기
for i in range(len(df)): 
    row = df.iloc[i]
    약물 = row['복용약물']
    dosage = row[약물 + '_용량']
    dosage_list.append(dosage)

# 2. 복용시간 리스트 만들기
for i in range(len(df)):
    row = df.iloc[i]
    약물 = row['복용약물']
    time = row[약물 + '_복용시간']
    time_list.append(time)

# 결과 출력
print(dosage_list)
print(time_list)
```

```python
df['용량'] = dosage_list
df['복용시간'] = time_list
df
```

```python
result = df[['환자ID', '복용약물', '용량', '복용시간']]
result
```
