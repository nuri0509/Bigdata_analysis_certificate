```python
import pandas as pd
import numpy as np

data = {
    'region' : ['north', 'south', 'east', 'west', 'north', 'south', 'east', 'west'],
    'product' : ['a', 'a', 'b', 'b', 'c', 'c', 'a', 'a'],
    'price' : [100, 150, 200, 130, 160, 120, 90, 110],
    'quantity' : [10, 15, 20, 13, 16, 12, 9, 11]
}
df = pd.DataFrame(data)
df
```

```python
# 1. groupby 설명
# 특정 컬럼을 기준으로 데이터를 그룹화하고, 각 그룹의 합계, 평균 등 기본 통계량을 계산할 수 있음
```

```python
df.groupby(['region','product'])[['price', 'quantity']].sum()
```

```python
# 그룹별 계산할 수 있는 여러가지 case

# count() : 각 그룹에서 NA가 아닌 값의 수
# nunique() : 각 그룹의 고유한 값의 수
# mean() : 각 그룹의 평균
# median() : 각 그룹의 중앙값
# max() : 각 그룹의 최대값
# min() : 각 그룹의 최소값
# sum() : 각 그룹의 합계
# std() : 각 그룹의 표준편차
# var() : 각 그룹의 분산
```

```python
# 2. 예제로 이해하기

# 예제1 - 그룹변수 1개, value 변수 1개, 요약방법 1개
# region별로 그룹화하여 price 합계 계산

result_1_1 = df.groupby('region')['price'].sum()  # 결과값 : 시리즈
result_1_2 = df.groupby('region')[['price']].sum()

print(type(result_1_1))
print(type(result_1_2))
```

```python
result_1_2
```

```python
# 예제2 - 그룹변수 2개, valud 변수 1개, 요약방법 1개
# region과 product별로 그룹화하여 price 합계 계산
result2 = df.groupby(['region', 'product'])[['price']].sum()
result2
```

```python
# 예제3 -  그룹변수 2개, value 변수 2개, 요약방법 2개
# region과 product별로 그룹화 후 price는 합계, quantity는 평균계산
result3 = df.groupby(['region', 'product'])[['price', 'quantity']].agg(['sum', 'mean'])
result3
```

```python
# 3. df.groupby 이후 필요한 전처리(index 초기화 방법)
# 1) 컬럼 이름 변경
# 2) index 초기화

result3.columns = ['price_sum', 'price_mean', 'quantity_sum', 'quantity_mean']
result3
```

```python
result3 = result3.reset_index()
result3
```

```python
# 4. 그룹별 결측치 확인 및 대체하기 (*중요 : 9회 기출)
import pandas as pd
import numpy as np

data = {
    '부서' : ['marketing'] * 9 + ['it'] * 9,
    '평가등급' : ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C',
              'A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    '근속연수' : [9, None, 11, 14, None, 16,19, None ,21,
              2, None, 4, 5, None, 7, 8, None,10]
}

# DataFrame 생성
df = pd.DataFrame(data)

df
```

```python
# 문제 : 근속연수의 결측치를 동일한 부서, 동일한 평가등급을 받은 그룹의 평균으로 대체하기
```

```python
# (부서, 평가등급) 조합별 근속연수 확인해보기
df.groupby(['부서', '평가등급'])[['근속연수']].mean()
```

```python
# (부서, 평가등급) 조합별 근속연수 평균으로 결측치 대체하기
df2 = df.copy()
df2.groupby(['부서', '평가등급'])['근속연수'].transform('mean')
# transform() 함수안에 'mean', 'median', 'max', 'min' 등으로 사용가능
```

```python
df2['근속연수'] = df['근속연수'].fillna(df.groupby(['부서','평가등급'])['근속연수'].transform('mean'))
df2
```
