```python
# 5.실전 예제풀이(lambda + apply + f-sting)
# 조건에 따른 컬럼 선택 및 값 추출

# 예제1 : 시험 유형에 따른 점수값 확인하기

## 데이터 생성 ##
import pandas as pd

df = pd.DataFrame({
    '시험유형' : ['A', 'B', 'A'],
    'A_점수' :  [70, 95, 88],
    'B_점수' : [70, 95, 88]
})
df
```

```python
# 유형 값에 알맞은 점수 선택하기(axis=1 넣기!)
df['선택된점수'] = df.apply(lambda x : x[f"{x['시험유형']}_점수"], axis=1)
df 
```

```python
# 코드 뜯어보기
# apply 함수에서 axis=1은 각 열마다 데이터 확인(동일 행 기준)
# x['시험유형'] : 각 행의 '시험유형' 컬럼 값을 가져온다. 첫 행에서는 A값을 가지고 옴
# f"{x['시험유형']}_점수" : x['시험유형']이 "A"면 -> "A_점수"가 된다.
# x[f"{x['시험유형']}_점수"] : 최종적으로 x['A_점수'] or x['B_점수']로 작동해서 해당 컬럼값을 가져오게 된다.
```

```python
# 예제2 : 학생이 선택한 과목의 점수와 강의실 위치를 확인해보기
```

```python

df = pd.DataFrame({
'학생ID': [201, 202, 203, 204, 205],
'선택과목': ['수학','영어','과학','수학','영어'],
'수학_점수': [88, 92, 85, 91, 87],
'영어_점수': [78, 81, 79, 75, 83],
'과학_점수': [89, 85, 90, 88, 84],
'수학_강의실': ['101호']*5,
'영어_강의실': ['102호']*5,
'과학_강의실': ['103호']*5
})
df
```

```python
# 예제2) 풀이1 : apply + lambda + f-string
```

```python
# 선택한 과목의 점수 확인하기
df['선택과목_점수'] = df.apply(lambda x : x[f"{x['선택과목']}_점수"], axis=1)
df
```

```python
# 선택한 과목의 강의실 확인하기
df['선택과목_강의실'] = df.apply(lambda row : row[f"{row['선택과목']}_강의실"], axis=1)
df
```

```python
# 최종 정리 : 학생별 선택과목의 점수 및 강의실 정보
result2_1 = df[['학생ID', '선택과목', '선택과목_점수', '선택과목_강의실']]
result2_1
```

```python
# 예제 2) 풀이 2: melt() + merge()

score = df.melt(
    id_vars=['학생ID', '선택과목'],
    value_vars=['수학_점수', '영어_점수', '과학_점수'],
    var_name ='과목',
    value_name='점수'
)
score
```

```python
score['과목'] = score['과목'].str.replace('_점수', '')
score
```

```python
room = df.melt(
    id_vars=['학생ID', '선택과목'],
    value_vars=['수학_강의실', '영어_강의실', '과학_강의실'],
    var_name='과목',
    value_name='강의실'
)
room
```

```python
room['과목'] = room['과목'].str.replace('_강의실', '')
```

```python
room
```

```python
result = pd.merge(score, room, on=['학생ID', '선택과목', '과목'])
result2_2 = result[result['선택과목'] == result['과목']]   # 핵심 아이디어
result2_2
```

```python
result2_2[['학생ID', '선택과목', '점수', '강의실']]
```

```python
# 위에 풀이1과 비교
result2_1
```

```python
# 예제3: 환자가 복용중인 약물의 용량과 복용시간 확인

### 데 이 터 생 성 ###
df = pd.DataFrame({
'환자ID': [301, 302, 303, 304, 305],
'복용약물': ['X','Y','Z','X','Y'],
'X_용량': [100, 150, 100, 120, 130],
'Y_용량': [200, 180, 190, 210, 170],
'Z_용량': [300, 310, 290, 305, 295],
'X_복용시간': ['아침']*5,
'Y_복용시간': ['점심']*5,
'Z_복용시간': ['저녁']*5
})
### 데 이 터 생 성 ###
df
```

```python
# 예제3) 풀이1 : apply + lambda + f-string

# 환자별 복용약물의 용량 확인하기
df['복용약물_용량'] = df.apply(lambda x : x[f"{x['복용약물']}_용량"], axis=1)
df
```

```python
# 환자별 복용약물의 복용시간 확인하기
df['복용약물_복용시간'] = df.apply(lambda x : x[f"{x['복용약물']}_복용시간"], axis=1)
df
```

```python
# 최종 정리 : 환자별 복용약물의 용량 및 복용시간 정보
result3_1 = df[['환자ID', '복용약물', '복용약물_용량', '복용약물_복용시간']]
result3_1
```

```python
# 예제 3) 풀이 2 : melt() + merge()
```

```python
dosage = df.melt(
    id_vars=['환자ID', '복용약물'],
    value_vars=['X_용량', 'Y_용량', 'Z_용량'],
    var_name='약물',
    value_name='용량'
)
dosage
```

```python
dosage['약물'] = dosage['약물'].str.replace('_용량', '')
dosage
```

```python
time = df.melt(
    id_vars=['환자ID', '복용약물'],
    value_vars=['X_복용시간', 'Y_복용시간', 'Z_복용시간'],
    var_name='약물',
    value_name='복용시간' 
)
time
```

```python
time['약물'] = time['약물'].str.replace('_복용시간', '')
time
```

```python
result = pd.merge(dosage, time, on=['환자ID', '복용약물', '약물'])
result = result[result['복용약물'] == result['약물']]
result
```

```python
result3_2 = result[['환자ID', '복용약물', '용량', '복용시간']]
result3_2
```
