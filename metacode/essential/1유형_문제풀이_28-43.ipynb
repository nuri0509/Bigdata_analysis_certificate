```python
# 문제 37
# mpg값이 25이상이면 high, 15이상 25 미만이면 medium, 15미만이면 low로 범주화하고
# mpg값이 high인 그룹과 low인 그룹의 disp 표준편차의 차이를 절대값으로 구하시오.

df = pd.read_csv('./data/mtcars.csv')
df.head()
```

```python
# (풀이 1)
# apply + lambda

df['mpg_level'] = df['mpg'].apply(lambda x : 'high' if x >= 25 else 'medium' if x >= 15 else 'low')
# print(df)

df1 = df[df['mpg_level'] == 'high']
result1 = df1['disp'].std()
print(result1)

df2 = df[df['mpg_level'] == 'low']
result2 = df2['disp'].std()
print(result2)

result3 = abs(result1 - result2)
print(round(result3, 1))
```

```python
# (풀이 2)
# List Comprehension(리스트 컴프리헨션)
df['mpg_level'] = ['high' if x >= 25 else 'medium' if x >= 15 else 'low' for x in df['mpg']]
df1 = df[df['mpg_level'] == 'high']
result1 = df1['disp'].std()

df2 = df[df['mpg_level'] == 'low']
result2 = df2['disp'].std()

result3 = abs(result1 - result2)
print(round(result3, 1))
```

```python
#(풀이 3)
# np.where
df['mpg_level'] = np.where(df['mpg'] >= 25, 'high',
                           np.where(df['mpg'] >= 15, 'medium', 'low'))

df1 = df[df['mpg_level'] == 'high']
result1 = df1['disp'].std()

df2 = df[df['mpg_level'] == 'low']
result2 = df2['disp'].std()

result3 = abs(result1 - result2)
print(round(result3, 1))
```

```python
# 문제 40
# 다음은 학생별(학생ID) 선택과목이 점수, 과목별 강의실 정보에 대한 데이터이다.
# 학점당 평균 점수가 가장 높은 학생의 학생 ID를 구하시오.
# * 학점당 평균 점수는 가중점수의 총합을 전체 학점수로 나눈 값이다.
# * 가중 점수는 과목별 점수x과목별 학점수로 계산한다.

## 복사 영역 ##
import pandas as pd
df = pd.DataFrame({
    '학생ID': [201, 202, 203, 204, 205],
   # '선택과목': ['수학', '영어', '과학', '수학', '영어'],
    '수학_점수': [88, 92, 85, 91, 87],
    '영어_점수': [78, 81, 79, 75, 83],
    '과학_점수': [89, 85, 90, 88, 84],
   # '수학_강의실': ['101호']*5,
   # '영어_강의실': ['102호']*5,
   # '과학_강의실': ['103호']*5
})
# 학점 정보 추가
df['수학_학점수'] = 3
df['영어_학점수'] = 2
df['과학_학점수'] = 1

## 복사 영역 ##
df
```

```python
# 파생변수 추가(가중변수의 총합)
df['total'] = df['수학_가중점수'] + df['영어_가중점수'] + df['과학_가중점수']
df['total1'] = df.loc[:, '수학_가중점수':'과학_가중점수'].sum(axis=1)
df
```

```python
df['전체학점수'] = df.loc[:, '수학_학점수':'과학_학점수'].sum(axis=1)
# print(df)

df['학점당평균'] = df['total'] / df['전체학점수']
df
# print(df)
```

```python
# 문제 41
# 다음은 어느 회사의 지역별(region) 성별에 따른 제품별 판매량 데이터이다.
# 지역별 성별에 따른 총 판매량을 구하고, 성별간 총 판매수 차이가 큰 지역의 지역명을 구하시오.
# (단, 총 판매수는 product1과 product2를 더한 값이다.)

## 복사 영역 ##
import pandas as pd
data = {
    'region': ['a']*4 + ['b']*4 + ['c']*4 + ['d']*4 + ['e']*4,
    'gender':  [1, 2, 1, 2,   2, 1, 2, 1,   1, 1, 2, 2,   2, 1, 1, 2,   1, 2, 2, 1],
    'product1': [
        1000, 2000, 2500, 3000,   # a
        4000, 120, 250, 350,      # b
        5000, 6000, 5500, 5800,   # c
        7000, 7200, 7100, 7300,   # d
        8000, 8200, 8100, 8300    # e
    ],
    'product2': [
        3000, 1500, 1500, 2000,   # a
        3500, 350, 290, 1000,     # b
        4000, 4500, 4200, 4300,   # c
        6500, 6700, 6600, 6800,   # d
        7500, 7700, 7600, 7800    # e
    ]
}
df = pd.DataFrame(data)
## 복사 영역 ##
df.head(10)
```

```python
# 문제 42
# 다음은 어느 회사의 지역별(region) 성별에 따른 제품별 판매량 데이터이다.
# product1 변수의 결측치를 각 region별 같은 성별의 평균 값으로 채우고,
# 지역별 product1과 product2 판매수의 차이가 가장 큰 지역을 찾으시오.

## 복사 영역 ##
import pandas as pd
data = {
    'region': ['a']*4 + ['b']*4 + ['c']*4 + ['d']*4 + ['e']*4,
    'gender':  [1, 2, 1, 2,   2, 1, 2, 1,   1, 1, 2, 2,   2, 1, 1, 2,   1, 2, 2, 1],
    'product1': [
        1000, 2000, 2500, 3000,   # a
        4000, 120, 250, 350,      # b
        5000, 6000, 5500, 5800,   # c
        7000, 7200, 7100, 7300,   # d
        8000, 8200, 8100, 8300    # e
    ],
    'product2': [
        3000, 1500, 1500, 2000,   # a
        3500, 350, 290, 1000,     # b
        4000, 4500, 4200, 4300,   # c
        6500, 6700, 6600, 6800,   # d
        7500, 7700, 7600, 7800    # e
    ]
}
df2 = pd.DataFrame(data)
df = df2.copy()
# 각 지역에서 첫 번째 등장하는 인덱스를 기준으로 결측치 처리
for region in ['a', 'b', 'c']:
    idx = df[df['region'] == region].index[0]
    df.loc[idx, 'product1'] = np.nan

## 복사 영역 ##
df.head(10)
```

```python
# (풀이)
df2 = df.copy()
df2['product1'] = df['product1'].fillna(df.groupby(['region', 'gender'])['product1'].transform('mean'))
df2.head(10)
```

```python
pivot1 = pd.pivot_table(df2,
                        index=['region'],
                        values=['product1', 'product2'],
                        aggfunc='sum')

pivot1
```

```python
pivot1['차이'] = abs(pivot1['product1'] - pivot1['product2'])
print(pivot1)

result = pivot1['차이'].idxmax()
print(result)  # 정답 : C (지역명)
```

```python
# 문제 43
# 다음은 연도별 과일의 판매량 및 재고 데이터이다.
# 각 연도별 최대 판매율(판매율 = sales/stock)을 가진 과일 종류를 찾아
# 연도별 해당 과일의 판매수(sales)를 모두 더하시오.

## 복사 영역 ##
df = pd.DataFrame({
    'year': [2022, 2022, 2023, 2023, 2024, 2024],
    'type': ['sales', 'stock'] * 3,
    'apple':  [50, 60,   45, 55,   40, 50],
    'banana': [30, 35,   25, 30,   20, 25],
    'orange': [70, 65,   80, 75,   90, 85],
    'grape':  [90, 85,   88, 83,   86, 81]})

## 복사 영역 ##
```

```python
# (풀이)
# type의 sales, stock 기준으로 필터링
df1 = df[df['type'] == 'sales']
df2 = df[df['type'] == 'stock']

print(df1)
print(df2)
```

```python
# 데이터를 'year'기준으로 병합
merged_df = pd.merge(df1, df2, on='year')  # suffixes=('_sales', '_stock')
merged_df
```

```python
# 각 과일마다 판매율 파생변수를 반복문으로 만들기
fruits = ['apple', 'banana', 'orange', 'grape']

for i in fruits :
    merged_df[f'{i}_판매율'] = merged_df[f'{i}_x'] / merged_df[f'{i}_y']

merged_df
```

```python
# 최대판매율 과일 찾기
merged_df['최대판매율과일'] = merged_df[[f'{i}_판매율' for i in fruits]].idxmax(axis=1)
merged_df['최대판매율과일2'] = merged_df.loc[:, 'apple_판매율':'grape_판매율'].idxmax(axis=1)
merged_df
```

```python
# 과일 유형만 따로 골라내기
merged_df['최대판매율과일'] = merged_df['최대판매율과일'].str.split('_').str.get(0)

# merged_df['최대판매율과일'] = merged_df['최대판매율과일'].str.replace('_판매율', '')
merged_df
```

```python
# 판매율이 가장 높은 과일 유형의 판매수 구하기
# 방법1 (apply + lambda)
merged_df['최종판매수'] = merged_df.apply(lambda x : x[f"{x['최대판매율과일']}_x"], axis=1)

# 방법2 (리스트 컴프리헨션)
merged_df['최종판매수1'] = [merged_df.loc[i, merged_df.loc[i, '최대판매율과일'] + '_x'] for i in range(len(merged_df))]

merged_df
```

```python
result = merged_df['최종판매수'].sum()
print(result)
```
