```python
# 문제 1.
# 다중회귀분석을 실시하고 각 물음에 답하시오.
# 독립변수 : age, sex, bmi
# 종속변수 : target

############### 실기환경 복사 영역 ###############
# 데이터 불러오기
import pandas as pd
import numpy as np
# 실기 시험 데이터셋으로 셋팅하기 (수정금지)
from sklearn.datasets import load_diabetes

# diabetes 데이터셋 로드
diabetes = load_diabetes()
x = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = pd.DataFrame(diabetes.target)
y.columns = ['target']
# 독립변수와 종속변수 설정
x = x[ ['age','sex','bmi'] ]
############### 실기환경 복사 영역 ###############

# 1-1) Rsq(결정계수) 값을 구하시오.
# 1-2) 회귀계수를 모두 구하시오.
# 1-3) 회귀식의 p-value 값을 구하시오.
```

```python
# 독립변수와 종속변수 설정

x = x[ ['age','sex','bmi'] ]
print(x.head())
print(y.head())
```

```python
# 모델링
from sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(x,y)
```

```python
# 1-1) Rsq(결정계수) 값을 구하시오.

model.score(x,y)
print(round(model.score(x,y),2))
```

```python
# 1-2) 회귀계수를 모두 구하시오.

print(np.round(model.coef_, 2))
```

```python
# 1-3) 회귀식의 p-value 값을 구하시오.

print(np.round(model.intercept_, 2))
```

```python
# 문제 2.
# 두 변수(bmi, target)에 대해 피어슨 상관분석을 실시하고 아래 물음에 답하시오.

############### 실기환경 복사 영역 ###############
# 데이터 불러오기
import pandas as pd
import numpy as np
# 실기 시험 데이터셋으로 셋팅하기 (수정금지)
from sklearn.datasets import load_diabetes
# diabetes 데이터셋 로드
diabetes = load_diabetes()
x = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y = pd.DataFrame(diabetes.target)
y.columns = ['target']
# 상관분석을 할 2가지 변수 설정
x = x['bmi']
y = y['target']
############### 실기환경 복사 영역 ###############

# 2-1) 귀무가설과 대립가설을 설정하시오.
# 2-2) 상관계수 값을 구하시오.
# 2-3) 검정통계량과 p-value 값을 구하시오
# 2-4) 귀무가설 채택/기각 여부를 결정하시오.
```

```python
# 2-1) 귀무가설과 대립가설을 설정하시오.

# 상관계수에 대한 검정실시

from scipy.stats import pearsonr
r, pvalue = pearsonr(x, y)

# H0 : 두 변수간 선형관계가 존재하지않는다( P = 0 )
# H1 : 두 변수간 선형관계가 존재한다( P != 0 )
```

```python
# 2-2) 상관계수 값을 구하시오.

print(round(r, 2))
print(round(pvalue, 2))
```

```python
# 2-3) 검정통계량과 p-value 값을 구하시오

# 통계량은 별도로 구해야 함
#  T = r * root(n-2) / root(1-r2) 
# r = 상관계수
# n = 데이터의 개수

n = len(x)
r2 = r**2
statistic = r * ((n-2)**0.5) / ((1-r2)**0.5)

print(round(statistic, 2))
```

```python
# 2-4) 귀무가설 채택/기각 여부를 결정하시오.
# 귀무가설 기각(대립가설 채택) - 두 변수간에 선형관계가 존재한다고 할수있다. 
```

```python
# 문제 3.
# 로지스틱회귀분석을 실시하고 각 물음에 답하시오.
# 독립변수 : sex, sibsp, fare
# 종속변수 : survived

############### 복사 영역 ###############
# 데이터 불러오기
import pandas as pd
import numpy as np
# Seaborn의 내장 타이타닉 데이터셋을 불러옵니다.
import seaborn as sns
df = sns.load_dataset('titanic')
# 분석 데이터 설정
df = df[ ['survived','sex','sibsp','fare'] ] # sex:성별, sibsp:탑승한 부모 및 자녀
# 성별을 map 함수를 활용해서 각각 1과 0에 할당한다.(여성을 1, 남성을 0)
# (실제 시험의 지시 조건에 따를 것)
df['sex'] = df['sex'].map({'female': 1,'male': 0 })
############### 복사 영역 ###############

# 3-1) 회귀계수를 모두 구하시오.
# 3-2) 로지스틱 회귀모형에서 sibsp 변수가 한단위 증가할 때
# 생존할 오즈가 몇 배 증가하는지 반올림하여 소수점 셋째 자리까지 구하시오.
# 3-3) 로지스틱 회귀모형에서 여성일 경우 남성에 비해 오즈가 몇 배 증가하는지
# 반올림하여 소수점 셋째 자리까지 구하시오.
```

```python
x = df.drop(['survived'], axis=1)
y = df['survived']
```

```python
from sklearn.linear_model import LogisticRegression

model1 = LogisticRegression(penalty=None)   ####페널티 default 값이 L2이므로 꼭 None 으로 지정해줘야 함.
model1.fit(x,y)
```

```python
# 3-1) 회귀계수를 모두 구하시오.

print(np.round(model1.coef_, 4))
print(np.round(model1.intercept_, 4))
```

```python
# 3-2) 로지스틱 회귀모형에서 sibsp 변수가 한단위 증가할 때
# 생존할 오즈가 몇 배 증가하는지 반올림하여 소수점 셋째 자리까지 구하시오.

# 회귀식 : P(1일 확률) = 1 / (1+exp(-f(x)))

result = np.exp(model1.coef_[0,1])   
print(round(result, 3))
```

```python
# 3-3) 로지스틱 회귀모형에서 여성일 경우 남성에 비해 오즈가 몇 배 증가하는지
# 반올림하여 소수점 셋째 자리까지 구하시오.

result2 = np.exp(model1.coef_[0,0])
print(round(result2, 3))
```
