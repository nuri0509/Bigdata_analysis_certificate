```python
'''

문제 1-1.
다음은 지역코드별 성별에 따라 정리된 개인 대출 데이터이다.
지역코드별 성별에 따른 총 대출액을 구하고,
성별간 총 대출액 차이가 큰 지역의 지역코드 값을 구하시오.
(단, 총 대출액은 대출액 a와 대출액 b를 더한 값이다.)

'''
```

```python
#### 데이터 복사영역(수정금지)####
import pandas as pd
import numpy as np
loan_values = {
    '지역1': {'대출A': [1000, 2000, 2500, 3000],
             '대출B': [3000, 1500, 1500, 2000]},
    '지역2': {'대출A': [4000, 120, 250, 350],
             '대출B': [3500, 350, 290, 1000]},
    '지역3': {'대출A': [5000, 6000, 5500, 5800],
             '대출B': [4000, 4500, 4200, 4300]},
    '지역4': {'대출A': [7000, 7200, 7100, 7300],
             '대출B': [6500, 6700, 6600, 6800]},
    '지역5': {'대출A': [8000, 8200, 8100, 8300],
             '대출B': [7500, 7700, 7600, 7800]},
}
# 지역 리스트와 각 지역의 성별 순서 (1, 1, 2, 2)
regions = ['지역1', '지역2', '지역3', '지역4', '지역5']
gender_sequence = [1, 1, 2, 2]

# 각 지역마다 반복하며 데이터를 구성
data_rows = []
for region in regions:
    for i, gender in enumerate(gender_sequence):
        data_rows.append({
            '지역코드': region,
            '성별': gender,
            '대출A': loan_values[region]['대출A'][i],
            '대출B': loan_values[region]['대출B'][i]
        })
df = pd.DataFrame(data_rows)
#### 데이터 복사영역(수정금지)####
```

```python
df.head(15)
```

```python
# 1. 지역코드별 성별에 따른 총 대출액을 구하고,
# 2. 성별간 총 대출액 차이가 큰 지역의 지역코드 값을 구하기.
# (단, 총 대출액은 대출액 a와 대출액 b를 더한 값이다.)
```

```python
df['총대출액'] = df['대출A'] + df['대출B']
df.head()
```

```python
g1 = df.groupby(['지역코드', '성별'])[['총대출액']].sum()
# *주의 
# [['총대출액']] : 대괄호 2개 -> 데이터프레임
# ['총대출액'] : 대괄호 1개 -> 시리즈타입
g1
```

```python
g2 = g1.reset_index()
g2
```

```python
# 풀이 1번 : pivot_table 활용!
pivot1= pd.pivot_table(df,
                             index='지역코드',
                             columns='성별',
                             values='총대출액')
pivot1
```

```python
pivot1.columns = ['성별1_총대출액', '성별2_총대출액']
pivot1
```

```python
pivot2 = pivot1.reset_index()
pivot2
```

```python
pivot2['차이'] = abs(pivot2['성별1_총대출액'] - pivot2['성별2_총대출액'])

# '차이' 값을 기준으로 내림차순 정렬
result = pivot2.sort_values(by ='차이', ascending=False)
print(result)
```

```python
# 풀이 2번 : filtering + merge 활용
g2
```

```python
# 성별별로 필터링
df1 = g2[g2['성별'] == 1]
df2 = g2[g2['성별'] == 2]
print(df1)
print(df2)
```

```python
df_total = pd.merge(df1, df2, on='지역코드')  # suffixes=('_1', '_2') 옵션은 선택
df_total
```

```python
df_total['차이'] = abs(df_total['총대출액_x'] - df_total['총대출액_y'])
df_total
```

```python
# '차이'값을 기준으로 내림차순 정렬
result2 = df_total.sort_values(by='차이', ascending=False)
print(result2)
```
