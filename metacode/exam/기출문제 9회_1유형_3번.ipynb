```python
'''
문제 1-3.
다음은 어떤 기업의 HR 관련 자료이다.
연봉/근속연수 값이 두번째로 높은 사람의 근속연수를 a,
연봉/고객만족도 값이 첫번째로 높은 사람의 근속연수를 b라고 할 때,
a+b 값을 구하시오.

'''

# 결측치 처리 조건
# 고객만족도 변수의 결측치는 모든 직원의 평균값으로,
# 근속연수 결측치는 같은 부서에서 평가등급이 동일한 직원의 평균값으로 대체하고 답을 구하시오.
# (단, 결측치 대체 시 소수점은 모두 버림으로 처리하시오.)
```

```python
#### 데이터 복사영역(수정금지)####
import pandas as pd
import numpy as np
import random

# 랜덤시드
random.seed(100)
np.random.seed(100)

# 설정
departments = ['Sales', 'Marketing', 'IT', 'Finance', 'HR']
grades = ['A', 'B', 'C']
n_rows = 100

# 데이터 생성
data = {
    '부서': np.random.choice(departments, n_rows),
    '평가등급': np.random.choice(grades, n_rows),
    '근속연수': np.random.choice(list(range(1, 21)) + [np.nan]*5, n_rows),  # 결측치 포함
    '연봉': np.random.randint(3000, 9000, n_rows),
    '교육참가횟수': np.random.randint(1, 6, n_rows),
    '고객만족도': np.random.choice(list(range(1, 6)) + [np.nan]*3, n_rows)  # 결측치 포함
}

df = pd.DataFrame(data)
#### 데이터 복사영역(수정금지)####
```

```python
df
```

```python
'''

연봉/근속연수 값이 두번째로 높은 사람의 근속연수를 a,           -  # 2번째 문제 조건
연봉/고객만족도 값이 첫번째로 높은 사람의 근속연수를 b라고 할 때,  -  # 3번째 문제 조건
a+b 값을 구하시오.


결측치 처리 조건  -   # 1번째 문제 조건 
고객만족도 변수의 결측치는 모든 직원의 평균값으로,
근속연수 결측치는 같은 부서에서 평가등급이 동일한 직원의 평균값으로 대체하고 답을 구하시오.
(단, 결측치 대체 시 소수점은 모두 버림으로 처리하시오.).                       

'''
```

```python
missing = df['고객만족도'].mean()
print(missing)
df['고객만족도'] = df['고객만족도'].fillna(int(missing))
df
```

```python
df['근속연수'] = df['근속연수'].astype(float)
```

```python
df['근속연수'] = df['근속연수'].fillna(df.groupby(['부서','평가등급'])['근속연수'].transform('mean'))
df
```

```python
df
```

```python
df['근속연수'] = df['근속연수'].astype(int)
df['고객만족도'] = df['고객만족도'].astype(int)
```

```python
df
```

```python
# 연봉/근속연수 값이 두번째로 높은 사람의 근속연수를 a,           -  # 2번째 문제 조건
# 연봉/고객만족도 값이 첫번째로 높은 사람의 근속연수를 b라고 할 때,  -  # 3번째 문제 조건
# a+b 값을 구하시오.
```

```python
df['연봉/근속연수'] = df['연봉'] / df['근속연수']
df['연봉/고객만족도'] = df['연봉'] / df['고객만족도']
df
```

```python
a = df.sort_values(by='연봉/근속연수', ascending=False)
print(a)
```

```python
# 연봉/근속연수 값이 두번째로 높은 사람의 근속연수 = a
a = a.iloc[2, :]
print(a)   # a = 1
```

```python
b = df.sort_values(by='연봉/고객만족도', ascending=False)
print(b)
```

```python
# 연봉/고객만족도 값이 첫번째로 높은 사람의 근속연수 =  b
b = b.iloc[0, :]
print(b)  # b = 4
```
