```python
# 1-1 신상데이터
# age컬럼에 대해 아래의 과정을 수행하고 (다)단계의 결과값을 출력하여라
# (가) 제 1사분위수와 제 3사분위수를 구하기
# (나) 두 값의 차이 절댓값 구하기
# (다) 그 값의 소수점 버리기
```

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/basic1_data04.csv')
print(df.head(10))
```

```python
q1 = df['age'].quantile(.25)
q3 = df['age'].quantile(.75)

print(int(abs(q3-q1)))
```

```python
# 1-2 facebook 데이터

# facebook 데이터셋을 불러와 (loves반응 + wows반응) / (reactions반응) 비율이 
# 0.4보다 크고 0.5 보다 작으면서, type이 video인 데이터개수를 구하시오.

df = pd.read_csv('https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/fb_data04.csv')
df.head(10)
```

```python
df['total'] = (df['loves']+df['wows']) / df['reactions']
df.head()
```

```python
cond1 = df['total'] > 0.4
cond2 = df['total'] < 0.5

cond3 = df[cond1&cond2]
cond3
```

```python
answer = cond3[cond3['type'] == 'video']
print(len(answer))
```
