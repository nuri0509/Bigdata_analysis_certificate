```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/boston.csv')
df.head()
```

```python
crim = df['CRIM'].sort_values(ascending=False)[:10]
print(crim)

tenth = crim.iloc[9]
print(tenth)
```

```python
import numpy as np

df['CRIM'] = np.where(df['CRIM']>=tenth ,tenth, df['CRIM'])
```

```python
# # 1-2 하우징 데이터
# 주어진 데이터의 첫번째 행부터 순서대로 80%까지의 데이터를 추출후 
# 'total_bedrooms'변수의 결측값을 'total_bedrooms'변수의 중앙값으로 대체하고,
# 대체전의 'total_bedrooms'변수의 표준편차 값과 대체후의 'total_bedrooms'변수 표준편차 값을 산출하여 
# 대체 전 표준편차값- 대체 후 표준편차 값을 구해 출력한다.
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/housing.csv')
df.head()
```

```python
per80 = int(len(df['total_bedrooms']) * 0.8)
print(per80)
```

```python
before = df.copy()
before = before.iloc[:per80,:]
print(before)
```

```python
after = before.copy()
median = after['total_bedrooms'].median()
after['total_bedrooms'] = after['total_bedrooms'].fillna(median)
```

```python
# after.isnull().sum()
```

```python
before = before['total_bedrooms'].std()
after = after['total_bedrooms'].std()

print(round(before-after, 3))
```

```python
# # 1-3.하우징 데이터
# 하우징 데이터에서 'latitude'컬럼의 이상치를 찾아 이상치들의 합을 산출하시오.
# 이상치 기준: 평균-(표준편차*1.5), 평균+(표준편차*1.5)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/housing.csv')
df.head()
df.tail()
```

```python
mean = df['latitude'].mean()
std = df['latitude'].std()

cond1 = df['latitude'] < mean - (std*1.5)
cond2 = df['latitude'] > mean + (std*1.5)

outlier = df[(cond1|cond2)]
outlier['latitude'].sum()
```
