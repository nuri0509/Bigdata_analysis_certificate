```python
# 1-1. 소방서 데이터

# 소방서별[출동시간-신고시간]의 평균을 구해, 가장 큰값을 찾고 그 값을 반올림하여 분단위로 도출하시오.

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/big_6_1_1.csv')
df.head()
```

```python
df.info()
```

```python
df['신고시간'] = pd.to_datetime(df['신고시간'])
df['출동시간'] = pd.to_datetime(df['출동시간'])
df.head()
```

```python
df['대기시간'] = df['출동시간'] - df['신고시간']
df.head()
```

```python
answer = df.groupby('소방서명')['대기시간'].mean()
print(answer)
```

```python
# 1-3. 범죄 발생 수 데이터

# 연도별로 총 범죄 건수(범죄유형별 발생건수의 합)의 월평균값을 구한 후, 그 값이 가장 큰 연도를 찾아,
# 해당연도의 총 범죄건수의 월평균값을 반올림하여 정수로 출력하세요.

import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Soyoung-Yoon/bigdata/main/big_6_1_3.csv')
df.head()
```

```python
df['년도'] = pd.to_datetime(df['월별']).dt.year
df['월'] = pd.to_datetime(df['월별']).dt.month
df.head()
```

```python
df['총범죄건수'] = df['강력범죄'] + df['교통범죄'] + df['사이버범죄'] + df['지능범죄'] + df['마약범죄'] + df['경범죄']
df.head()
```

```python
monthly = df.groupby('년도')['총범죄건수'].sum() / 12
# monthly
print(round(monthly.max()))
```
